```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Give Colab access to Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Create file path
filepath = '/content/drive/MyDrive/Elwha project/SSC and river discharge/Elwha_DailySedimentLoads_2011to2020.csv'

# Load csv file
df = pd.read_csv(filepath)
df['Date'] = pd.to_datetime(df['Date'])
df.index = df['Date']
df.drop(columns=['Date'], inplace = True)
display(df)
```

```python
# Plot SSC, discharge, etc. using same x-axis for all subplots

# Create figure
fig, (ax1,ax2) = plt.subplots(nrows=2, sharex=True, figsize=(20, 10))

# Daily Discharge
ax1.set(title='Elwha River', 
        ylabel='Daily Discharge (m3/s)')
ax1.scatter(x=df.index, y=df['Daily Discharge (m3/s)'], s=5)
# Daily SSC
ax2.set(xlabel='Date',
       ylabel='Daily SSC (mg/L)')
ax2.scatter(x=df.index, y=df['Daily SSC (mg/L)'], s=5, c='r')
```

```python
# Create figure
fig, (ax1,ax2,ax3) = plt.subplots(nrows=3, sharex=True, figsize=(20, 12))

# Total sediment discharge (tonnes)
ax1.set(title='Elwha River', 
        ylabel='Total sediment discharge (tonnes)')
ax1.scatter(x=df.index, y=df['Total sediment discharge (tonnes)'], s=3)

# Estimated daily ungauged bedload (tonnes)
ax2.set(ylabel='Estimated daily ungauged bedload (tonnes)')
ax2.scatter(x=df.index, y=df['Estimated daily ungauged bedload (tonnes)'], s=3, c='r')

# Daily Suspended-Sediment Load (tonnes)
ax3.set(xlabel='Date',
       ylabel='Daily Suspended-Sediment Load (tonnes)')
ax3.scatter(x=df.index, y=df['Daily Suspended-Sediment Load (tonnes)'], s=3, c='magenta')
```
